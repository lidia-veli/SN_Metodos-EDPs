## Ejercicio 2
$0<x<1, 0<y<1$  
  
$
\begin{cases}
    A(u) = 0 \\
    u(x,0) = 0 \\
    u(x,1) = x^2 \\
    u(0,y) = 1-y^2 \\
    u(1,y) = 1
\end{cases}
$

```python
# condiciones iniciales
a = 0
b = 1
c = 0
d = 1
N = 20
M = 20

# Paso
h = (b-a)/N  # en eje x
k = (d-c)/M  # en eje y

# Inicializar la matriz w
w = [[0 for _ in range(M+1)] for _ in range(N+1)]


# LAPLACIANO
def f_lapl(i, j):
    return 0

# Condiciones de frontera
for i in range(N+1):
    x = a + i*h
    w[i][0] = 0
    w[i][M] = x**2

for j in range(M+1):
    y = c + j*k
    w[0][j] = 1 - y**2
    w[N][j] = 1


# -------------------------------------------------------------------
# LLAMADA METODO -----------------------------------------
# -------------------------------------------------------------------
    
gauss_seidel_elipt_poisson(a,b,c,d, N,M, h,k, w, f_lapl)
```

## Ejercicio 3
$0<x<1, 0<y<1$  
  
$
\begin{cases}
    A(u) = xy(1-x)(1-y) \\
    u(x,0) = 0 \\
    u(x,1) = 0 \\
    u(0,y) = 0 \\
    u(1,y) = 0
\end{cases}
$

```python
# condiciondes iniciales
a = 0
b = 1
c = 0
d = 1
N = 20
M = 20

# paso
h = (b-a)/N  # en eje x
k = (d-c)/M  # en eje y

# inicializar la matriz w
w = [[0 for _ in range(M+1)] for _ in range(N+1)]


# LAPLACIANO
def f_lapl(i, j):
    x = a + i*h
    y = c + j*k

    return x*y*(1-x)*(1-y)

# Condiciones
for i in range(N+1):
    w[i][0] = 0
    w[i][M] = 0

for j in range(M+1):
    w[0][j] = 0
    w[N][j] = 0


# -------------------------------------------------------------------
# LLAMADA METODO -----------------------------------------
# -------------------------------------------------------------------
    
gauss_seidel_elipt_poisson(a,b,c,d, N,M, h,k, w, f_lapl)
```
